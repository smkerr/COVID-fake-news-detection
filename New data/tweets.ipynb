```python
import os
import tweepy as tw
import pandas as pd
from twitter_authentication import consumer_key
from twitter_authentication import consumer_secret
from twitter_authentication import access_token
from twitter_authentication import access_token_secret
```

```python
auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth, wait_on_rate_limit=True)
```

```python
#filter for user and then manually detect tweets that are not about covid and not english
# get tweets from specific user1

user = "WHO"
limit = 100

tweets = api.user_timeline(screen_name=user, count=limit, tweet_mode="extended", since_id=1477226298560950276, exclude_replies=True, include_rts=False)

for tweet in tweets:
  print(tweet.full_text)
```

```python
columns = ['Tweet']
data = []

for tweet in tweets:
    data.append([tweet.full_text])

df = pd.DataFrame(data, columns=columns)

print(df)
```
