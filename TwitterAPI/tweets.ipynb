```python
import os
import tweepy as tw
import pandas as pd
from twitter_authentication import consumer_key
from twitter_authentication import consumer_secret
from twitter_authentication import access_token
from twitter_authentication import access_token_secret
```

```python
auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth, wait_on_rate_limit=True)
```

```python
# Define the search term and the date_since date as variables
search_words = "from:WHO"
date_since = "2022-04-01"
limit= 5
```

```python
# Collect tweets
tweets = tw.Cursor(api.search,
              q=search_words,
              lang="en",
              since=date_since).items(limit)

# Iterate and print tweets
for tweet in tweets:
    print(tweet.text)
```

```python
# get tweets from specific user

user = "WHO"
limit = 10

tweets = api.user_timeline(screen_name=user, count=limit, tweet_mode="extended")

for tweet in tweets:
  print(tweet.full_text)

# How to combine user and hashtag?
# tweepy.Cursor(api.search,q="from:twitter #ForTheThrone") doesn't give any output
# possible solution: first scrape timeline of specific user and then filter for hashtag covid?

```

```python
# create DataFrame
columns = ['User', 'Tweet']
data = []

for tweet in tweets:
    data.append([tweet.user.screen_name, tweet.full_text])

df = pd.DataFrame(data, columns=columns)

print(df)
```
