```python
#Import the dependencies
from bs4 import BeautifulSoup
import pandas as pd
import requests
import urllib.request
import time
import numpy
```

```python
#Create lists to store the scraped data

dates = []
statements = []
dates2 = []
statements2 = []
```

```python
#Create a function to scrape the site for fake news related to covid
def scrape_website_fake(page_number):
    page_num = str(page_number) #Convert the page number to a string


    URL = 'https://www.politifact.com/factchecks/list/?page={}&category=coronavirus&ruling=false'.format(page_num)


    webpage = requests.get(URL)  #Make a request to the website
    time.sleep(3)
    soup = BeautifulSoup(webpage.text, "html.parser") #Parse the text from the website
    #Get the tags and it's class
    statement_footer =  soup.find_all('footer',attrs={'class':'m-statement__footer'})  #Get author and date
    statement_quote = soup.find_all('div', attrs={'class':'m-statement__quote'}) #Get the statement
    for i in statement_footer:
        link1 = i.text.strip()
        name_and_date = link1.split()
        month = name_and_date[4]
        day = name_and_date[5]
        year = name_and_date[6]
        date = month+' '+day+' '+year
        dates.append(date)
    #Loop through the div m-statement__quote
    for i in statement_quote:
        link2 = i.find_all('a')
        statements.append(link2[0].text.strip())

#create function to scrape the site for pants on fire news related to covid

def scrape_website_pants(page_number):
    page_num = str(page_number) #Convert the page number to a string


    URL = 'https://www.politifact.com/factchecks/list/?page={}&category=coronavirus&ruling=pants-fire'.format(page_num)


    webpage = requests.get(URL)  #Make a request to the website
    time.sleep(3)
    soup = BeautifulSoup(webpage.text, "html.parser") #Parse the text from the website
    #Get the tags and it's class
    statement_footer =  soup.find_all('footer',attrs={'class':'m-statement__footer'})  #Get author and date
    statement_quote = soup.find_all('div', attrs={'class':'m-statement__quote'}) #Get the statement
    for i in statement_footer:
        link1 = i.text.strip()
        name_and_date = link1.split()
        month = name_and_date[4]
        day = name_and_date[5]
        year = name_and_date[6]
        date = month+' '+day+' '+year
        dates2.append(date)
    #Loop through the div m-statement__quote
    for i in statement_quote:
        link2 = i.find_all('a')
        statements2.append(link2[0].text.strip())
```

```python
n=5
for i in range(1, n):
    scrape_website_fake(i)
```

```python
#Create dataFrame fake
FN = pd.DataFrame(columns = ['statement', 'date'])
FN['statement'] = statements
FN['date'] = dates
#Show the data set
print(FN)
```

```python
#Loop through 'n-1' webpages to scrape the data false
n=2
for i in range(1, n):
    scrape_website_pants(i)
```

```python
#Loop through 'n-1' webpages to scrape the data pants-fire
#Create dataFrame pants-fire
PFN = pd.DataFrame(columns = ['statement', 'date'])
PFN['statement'] = statements2
PFN['date'] = dates2
#Show the data set
print(PFN)
```

```python
#combine Fake news and pants on fire news
frames = [FN, PFN]

result = pd.concat(frames)

print(result)
```

```python
df = result.loc[result.date.str.contains('January|February|March|April'),:]
print(df)
```

```python
df = df.drop(columns=["date"])
```
